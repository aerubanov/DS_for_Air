```python
import pandas as pd
import matplotlib.pyplot as plt
import os

%matplotlib notebook
```

```python
DATA_FOLDER = '../DATA'
files = os.listdir(DATA_FOLDER)
```

```python
start_dates = []
end_dates = []
for f in files:
    try:
        data = pd.read_csv(os.path.join(
            DATA_FOLDER, f), delimiter=';', parse_dates=['timestamp'])
        start_dates.append(data.timestamp.min())
        end_dates.append(data.timestamp.max)
    except ValueError:
        print(f)
start_dates, end_dates
```
